```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
jumsu = [49,62,74,82,96,100]
hour = [1,2,3,4,5,6]
cnt = [0,0,1,2,2,1]
dataDF = pd.DataFrame({"jumsu":jumsu,"cnt":cnt,"hour":hour})
dataDF
```

```python
# 시간과 점수와의 관계확인
import matplotlib.pyplot as plt
plt.plot(dataDF.hour,dataDF.jumsu)
```

```python
dataDF.corr()
```

```python
plt.hist(dataDF.jumsu,bins=4)
```

```python
dataDF['jumsu'].plot(kind='hist',bins=4,density=True)
```

```python
# 학습 방법 선정
# 독립 & 종속 변수 관계확인
# 데이터 + 라벨 => 지도학습
# 점수를 예측하기 ==> 회귀(Regreesion)
```

```python
model = LinearRegression()
```

```python
X = dataDF.hour.to_numpy().reshape(-1,1) #2차원으로 만들어!
y = dataDF.jumsu #타깃
type(X)
```

```python
X.shape, y.shape
```

```python
model.fit(X,y)
```

```python
model.coef_ , model.intercept_ #계수 / 절편
```

```python
y_pre = model.predict(X)
# jumsu = [49,62,74,82,96,100] 
```

```python
model.score(X,y)

#그렇다면 어떻게 score가 확인되는지
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y,y_pre)
# 얼마나 오차랑 차이가 나는가?
```
