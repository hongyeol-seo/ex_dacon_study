```python
train_x=train.drop(['Age','Cabin','Fare','Embarked','PassengerId','Name','Ticket'], axis=1)
train_y=train['Survived']
test_x=test.drop(['Age','Cabin','Fare','Embarked','PassengerId','Name','Ticket'], axis=1)
```

```python
#Pclass 
#Sex
#Sibsp 함께 탐승한 형제자매, 아내, 남편의 수
#Parch : 함께 탐승한 부모, 자식의 수
train_x.head(3)
```

```python

#Sibsp 함께 탐승한 형제자매, 아내, 남편의 수
train_x['SibSp'] = train_x['SibSp'].apply(lambda x : 1 if x >= 1 else 0)
test_x['SibSp'] = test_x['SibSp'].apply(lambda x : 1 if x >= 1 else 0)

#Parch : 함께 탐승한 부모, 자식의 수
train_x['Parch'] = train_x['Parch'].apply(lambda x : 1 if x >= 1 else 0)
test_x['Parch'] = test_x['Parch'].apply(lambda x : 1 if x >= 1 else 0)

#남자, 여자
train_x['Sex'] = train_x['Sex'].apply(lambda x : 1 if x == 'male' else 0)
test_x['Sex'] = test_x['Sex'].apply(lambda x : 1 if x == 'male' else 0)

train.head(3)
```
